```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_training_performance(file_path, metric="Average Score"):
    """
    Plots training performance over time based on the specified metric using Seaborn.

    Parameters:
        file_path (str): Path to the Excel file.
        metric (str): The metric to plot, either "Average Score" or "Average Win Rate".
    """
    # Load the Excel file
    df = pd.read_excel(file_path)
    
    # Check if the metric is valid
    if metric not in df.columns:
        print(f"Error: {metric} not found in the data.")
        print("Please choose either 'Average Score' or 'Average Win Rate'.")
        return
    
    # Set up the Seaborn plot style
    sns.set(style="whitegrid")
    
    # Plotting the data
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x='Number of Training Games', y=metric, marker='o', color='b', label=metric)
    
    # Adding labels and title
    plt.xlabel("Number of Training Games")
    plt.ylabel(metric)
    plt.title(f"Training Performance Over Time - {metric}")
    
    # Annotate each data point with its value
    # for i, value in enumerate(df[metric]):
    #     plt.text(df['Number of Training Games'][i], value, f"{value:.2f}", ha='center', va='bottom')

    # Show legend and plot
    plt.legend()
    plt.show()
```

```python
plot_training_performance("results.xlsx", metric="Average Score")
```

```python
def plot_scores_and_actions(file_path, columns_to_plot):
    """
    Plots a line plot for specified columns (e.g., Scores and Action) from an Excel file.

    Parameters:
        file_path (str): Path to the Excel file.
        columns_to_plot (list): List of columns to plot, such as ['Score', 'Action'].
    """
    # Load the Excel file
    df = pd.read_excel(file_path)
    
    # Drop rows with blank cells in the specified columns to plot
    df = df.dropna(subset=columns_to_plot)
    
    # Set up the Seaborn plot style
    sns.set(style="whitegrid")
    
    # Plot each specified column as a line plot
    plt.figure(figsize=(12, 6))
    for column in columns_to_plot:
        sns.lineplot(data=df, x=df.index, y=column, label=column, marker='o')
    
    # Adding labels and title
    plt.xlabel("Training Episode")
    plt.ylabel("Values")
    plt.title("Scores and Actions Over Time")
    
    # Adding a legend
    plt.legend()
    
    # Show the plot
    plt.show()
```

```python
plot_scores_and_actions("data_pacman_sarsa.xlsx", columns_to_plot=["Scores", "Actions"])
```
